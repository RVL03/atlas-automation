```python
## Filter value only desired index
landsat_ndvi = df_landsat_data[df_landsat_data["index"] == "ndvi"]
landsat_ndvi
```

```python
# Plot paramaeters setting
dpi = 100
linewidth = 0.7
fontdict = {'fontsize': 15,
            'fontweight':"bold"}
pad = 15
loc = "center"
```

```python
## PLot the value 
fig, ax = plt.subplots(figsize=(5,5),dpi=dpi)
ax.set_title('Landsat Mean NDVI 2013-2022 (Jurisdiction vs Project)',
             fontdict=fontdict,
             pad=pad,
             loc=loc)
ax.grid(True)
sns.set_style("ticks")
sns.barplot(data=landsat_ndvi, 
            x="year",
            y="mean",
            hue="scale",
            palette=["lightgreen","#0ea98b"],
            errorbar=None)
ax.set_xlabel("Year",fontsize=16)
ax.set_ylabel("NDVI",fontsize=16)
plt.legend(bbox_to_anchor=(0.5, -0.27), loc='lower center', borderaxespad=0)
plt.savefig('GOR_CHRT01_JvsP_Landsat_NDVI_2013-2022.png',bbox_inches ='tight')
```
